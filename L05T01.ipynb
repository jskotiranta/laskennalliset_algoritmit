```python
#Etsitään funktion minimi kätevästi laskentahilalla
```

```python
import numpy as np
import matplotlib.pyplot as plt #tarvitaan jos halutaan grafiikkaa
```

```python
delta = 0.05
range_xmin = 0.0
range_xmax = 25.0
range_ymin = 0.0
range_ymax = 25.0
```

```python
#luodaan laskentahila ylläolevilla spekseillä
x = np.arange(range_xmin,range_xmax,delta)
y = np.arange(range_ymin,range_xmax,delta)
X, Y = np.meshgrid(x, y)
```

```python
#funktio, jonka ääriarvoista ollaan kiinnostuneita
f = 2*(X - 18)**2 + X*Y + (Y - 20)**2
```

```python
#katsotaan, näyttääkö funktio järkevältä
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, f)
plt.show()
```

```python
#poimitaan indeksi, jossa minimi sijaitsee ja tulostetaan vastaavat koordinaatit
ind = np.unravel_index(np.argmin(f,axis=None),f.shape)
print("Minimi on kohdassa (",round(X[ind],1),",",round(Y[ind],1),")")
```
